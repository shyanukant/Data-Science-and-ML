### 2. Ordinal Encoding 

Convert categorical data into numerical bassed on ranking.

```python
import datetime
today = datetime.datetime.today()
```

```python
date = [today - datetime.timedelta(i) for i in range(10)]
date
```

```python
data = pd.DataFrame(date, columns=['date'])
data
```

```python
data['day'] = data['date'].dt.day_name()
data
```

```python
from calendar import day_name
dict = {}
for rank, value in enumerate(day_name):
    dict[value]= rank
dict
```

```python
data['ordinal'] = data['day'].map(dict)
data
```

```python
# using sklean
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
```

```python
# reset dataset
data2 = data.drop('ordinal', axis=1)
data2
```

```python
encoder.fit(data2)
encoder.transform(data2)
```

### 3. Count and frequency Encoding

Count the frequency of each category then replace with that number(frequency)

```python
df = pd.read_csv('adult-training.csv')
df.head()
```

```python
df.drop(['39',  ' 77516',' 13', ' 2174', ' 0', ' 40',' <=50K'], axis=1, inplace=True)
df.head()
```

```python
columns = ['employee_status', 'education', 'status', 'degisnation','Familiy_role','race','gender','country']
df.columns=columns
df.head()
```

```python
df['country'].unique()
```

```python
# count, how many categories are available in each column
for i in df.columns:
    print(i, ':', len(df[i].unique()),'Category\n')
```

```python
df['country'].value_counts().head(10)
```

```python
# choose one- country
country_cate = df['country'].value_counts().to_dict()
# replace each country with their count(how much time it repeat)
df['country'] = df['country'].map(country_cate)
df.head()
```

### Advantange
Not increasing feature space

### Disadvantage
It will provide same weight if the frequencies are same of more than one category

## 4. Target Guided Ordinal Encoding

1. Ording the labels according to the target(calculate mean of category)
2. Replace the labels by the joint probability of being 1 or 0

```python
df= pd.read_csv('titanic.csv', usecols=['Survived','Cabin'])
df.head()
```

```python
df['Cabin'].fillna('Missing', inplace=True)
```

```python
df.head()
```

```python
# take only first alphabet of cabin's column
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head(10)
```

```python
df.Cabin.unique()
```

```python
# find mean of categories
df.groupby(['Cabin'])['Survived'].mean()
```

```python
# sort the value
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
# take only index of cabin
label = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
label
```

```python
ordinal_label = {k:i for i, k in enumerate(label)}
ordinal_label
```

```python
df['Cabin_ordinal'] = df['Cabin'].map(ordinal_label)
df.head(10)
```

## 5. Mean Encoding

```python
# calculate mean and make dictionary
mean_label = df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_label
```

```python
df['mean_encode']= df['Cabin'].map(mean_label)
df.head(10)
```

### Advantages
1. its make some monotonic relation between features and target
2. capture information within the label, therefore more prodectibe features


#### Disadvantage
its leads to overfitting

## 6. Probability Radio Encoding

its similler to mean ordinal encoding

```python
# sort the value
prod = df.groupby(['Cabin'])['Survived'].mean().sort_values()
prod
```

```python
# probability of person alive
prod_df = pd.DataFrame(prod)
prod_df
```

```python
# probability of person died
prod_df['Died'] =  1- prod_df['Survived']
prod_df
```

```python
prod_df['Probability_ratio'] = prod_df['Survived']/prod_df['Died']
prod_df
```

```python
# we using probability_ratio column into our dataset
prod_encoded = prod_df['Probability_ratio'].to_dict()
df['probability_encoded'] = df['Cabin'].map(prod_encoded)
```

```python
df.head(10)
```

1. Probability of Survived based on cabin ---- Categorical Feature
2. Probability of Not Survived -->b 1-probability of Survived
3. Ratio = prod(Survived)/prod(Not Survived)
4. Dictionary to map cabin with probability
5. Replace with the categorical feature
